```python
import torch
import numpy as np
```

```python
shape = (3,2)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f'Random Tensor: \n {rand_tensor} \n')
print(f'Ones Tensor: \n {ones_tensor} \n')
print(f'Zeros Tensor: \n {zeros_tensor}')
```

```python
tensor = torch.rand((3,2), device='cuda')
print(f"shape: {tensor.shape}")
print(f"dtype: {tensor.dtype}")
print(f"device: {tensor.device}")
```

```python
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else 'cpu'
tensor = tensor.to(device)
tensor
```

```python
tensor = torch.ones(4,4)
tensor[:,1] = 0
tensor
```

```python
print(tensor.T)
```

```python
print(tensor @ tensor.T)
```

### Bridge with NumPy
Tensors on the CPU and NumPy arrays can share their underlying memory locations, and changing one will change the other.

```python
t = torch.ones(5)
print(f't: {t}')
n = t.numpy()
print(f'n: {n}')
```

Numpy can only handle cpu tensors.(Cannot handle gpu tensors). If we create the tensor on the gpu and try to make a numpy array by referencing that, it would return an error. We first need to move it to the cpu and do the operation

```python
t.add_(2)
print(f"t: {t}")
print(f"n: {n}")
```

#### Numpy Array to Tensor
Here also it's like a bridge changes in the numpy array will also reflect in the tensor.

```python
n = np.ones(5)
t = torch.from_numpy(n)
t
```

```python
np.add(n, 2, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
