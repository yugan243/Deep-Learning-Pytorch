```python
import torch
```

```python
#data points
X = torch.tensor([-0.6, -0.4, -0.2, 0.0, 0.2, 0.4])
y = torch.tensor([-0.1, -0.22, 0.6, 0.95, 1.26, 1.72])

```

```python
# Linear Regression model parameters
W = torch.rand(1, requires_grad=True)
b = torch.rand(1, requires_grad=True)
print(f"W: {W.item()}, b: {b.item()}")
```

```python
# Regressor model
# y_hat = W*X + b

lr = 0.01

for step in range(5000):
  if W.grad != None:
    W.grad.zero_()
  if b.grad != None:
    b.grad.zero_()

  y_hat = W * X + b
  mse = ((y_hat - y)**2).sum()
  mse.backward()

  with torch.no_grad():
    W -= lr * W.grad
    b -= lr * b.grad
  if step% 100 == 0:
    print(f"W: {W.item()}, b: {b.item()}, loss: {mse}")
```

## Swap the entire process with a nn.Linear (Linear layer)


```python
import torch.nn as nn
import torch.optim as optim
```

```python
# data points
X = torch.tensor([-0.6, -0.4, -0.2, 0.0, 0.2, 0.4])
y = torch.tensor([-0.1, -0.22, 0.6, 0.95, 1.26, 1.72])

```

nn.Linear expects shape (n_samples, n_features). so we are reshaping it.

```python
X = X.view(-1, 1)
y = y.view(-1, 1)
```

```python
model = nn.Linear(in_features=1, out_features=1)
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

epochs = 5000

for epoch in range(epochs):
  # forward pass
  y_pred = model(X)
  loss = criterion(y_pred, y)

  # backward pass
  optimizer.zero_grad()   # replaces your manual x.grad.zero_()
  loss.backward()         # same as before — autograd computes gradients

  # update weights
  optimizer.step()        # replaces your manual `x -= lr * x.grad`

  if epoch % 100 == 0:
    print(f"Epoch: {epoch}, Loss: {loss.item()}")

```

```python
W, b = model.weight.item(), model.bias.item()
print(f"\nLearned: y = {W} * X + {b} ")
```

### V3: Wrap it as a proper nn.Module subclass

```python
class LinearRegressionModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(1, 1)


  def forward(self, X):
    return self.linear(X)

model = LinearRegressionModel()

X = torch.tensor([[-0.6], [-0.4], [-0.2], [0.0], [0.2], [0.4]])
y = torch.tensor([[-0.1], [-0.22], [0.6], [0.95], [1.26], [1.72]])

criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

for epoch in range(5000):
  y_pred = model(X)
  loss = criterion(y_pred, y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch%100 == 0:
    print(f"epoch {epoch}: loss = {loss.item():.4f}")
```

```python
W, b = model.linear.weight.item(), model.linear.bias.item()
print(f"\nLearned: y = {W} * X + {b} ")
```
