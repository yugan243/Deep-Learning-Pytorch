```python
import torch
from torchvision.models import resnet18, ResNet18_Weights
```

```python
model = resnet18(weights=ResNet18_Weights.DEFAULT)
data = torch.rand(1, 3, 64, 64)
labels = torch.rand(1, 1000)
```

```python
model
```

```python
prediction = model(data) #forward pass
```

Now we calculate the models error using it's prediction and the corresponding label. and we call the  **.backward()** on that error tensor. Autograd then calculates  and stores the gradients for each model parameter in the parameter's **.grad** attribute.

```python
loss = (prediction - labels).sum()
print(loss)
loss.backward() #backward pass
```

Now we load an optimizer, in this case it is SGD with a learning rate of 0.02 and momentum of 0.9. We register all the parameters of the model in the optimizer.

```python
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
```

Finally, we call the **.step()** to initiate the gradient decent. The optimizer will then adjusts each parameters by its gradient stored in **.grad**.

```python
optim.step() #gradient decent
```

```python
import torch

x = torch.rand(3, requires_grad=True)
print(x)

y = x+2
print(y)

z = y*y*2
z = z.mean()
print(z)

z.backward() # dz/dx
print(x.grad)
```
