```python
x = torch.tensor([
    [1,2,3],[4,5,6]
])
x
```

This tensor is stored contiguosly in memory: 1, 2, 3, 4, 5, 6. Only the interpreting/visible order looks like this  
1 2  3
      
4 5  6


```python
x.view(6)
```

This works because the visible order and the memory order matches.


```python
y = x.t()
y
```

Pytorch usually does not moves the values in the memory. Memory is still 1, 2, 3, 4, 5, 6. But the logical order of the y is 1, 4, 2, 5, 3, 6.

```python
y.view(6)
```

This error throws because the existing physical memory order and the flatterned transpose order does not match. This is why reshape comes into play

Reshape will use view if possible. if not it would create a copy of the orginal data and rearrange it.

```python
y.reshape(6)
```
