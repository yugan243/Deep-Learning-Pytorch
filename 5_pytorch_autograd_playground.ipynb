```python
import torch
```

```python
x = torch.tensor([
    [2.0, 2.0],
    [4.0, 8.0]
], requires_grad=True)

y = torch.tensor([
    [1.0, 1.0],
    [3.0, 2.0]
], requires_grad=True)
```

```python
print(x)
print(x.dtype)
print(y)
print(y.dtype)
```

```python
z = (3*x**2 + 5*y**2).sum()
```

```python
z.backward()
```

```python
x = torch.tensor([[1, 2], [3, 4]], requires_grad=True, dtype=torch.float32)
x
```

```python
for i in range(3):
  if x.grad != None:
    x.grad.zero_()
  y = (x**2).sum()
  y.backward()
  print(x.grad)
```

```python
x = torch.tensor([[1, 2], [3, 4]], requires_grad=True, dtype=torch.float32)

with torch.no_grad():
  y = x * 2
print(y.requires_grad)
```

Here pytorch doesn't build the computational graph.

```python
x = torch.tensor(5.0, requires_grad=True)
lr = 0.1

for step in range(10):
  y = (x - 3) ** 2
  y.backward()

  with torch.no_grad():
    x -= lr * x.grad
  x.grad.zero_()

  print(f"setp: {step}, x: {x.item()}, loss: {y.item()}")
```

Without **no_grad()**, PyTorch would try to remember:

`new x = old x - learning_rate × gradient`

as another operation in the computation graph.

But we do not want to later calculate derivatives through the update rule itself. We only want derivatives of the loss with respect to the current value of x.

So we want:



*   Loss calculation → tracked
*   Gradient calculation → tracked
*   Parameter update → not tracked
